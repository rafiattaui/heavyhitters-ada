```python
import csv
import matplotlib.pyplot as plt

limits = []
runtime = []
memory = []

# Read CSV
with open("cms.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        limits.append(int(row["limit"]))
        runtime.append(float(row["cms_runtime"]))
        memory.append(float(row["cms_memory"]))

# Plot runtime growth
plt.figure()
plt.plot(limits, runtime, marker='o')
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Runtime (seconds)")
plt.title("CMS Runtime Growth")
plt.show()

# Plot memory usage
plt.figure()
plt.plot(limits, memory, marker='o')
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Memory Usage (KB)")
plt.title("CMS Memory Usage Growth")
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

limits = []

mg_runtime = []
mg_memory = []
mg_abs_error = []
mg_rel_error = []

lc_runtime = []
lc_memory = []
lc_abs_error = []
lc_rel_error = []

# Read CSV
with open("metrics.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        limits.append(int(row["limit"]))

        mg_runtime.append(float(row["mg_runtime"]))
        mg_memory.append(float(row["mg_memory_kb"]))
        mg_abs_error.append(float(row["mg_avg_abs_error"]))
        mg_rel_error.append(float(row["mg_avg_rel_error"]))

        lc_runtime.append(float(row["lc_runtime"]))
        lc_memory.append(float(row["lc_memory_kb"]))
        lc_abs_error.append(float(row["lc_avg_abs_error"]))
        lc_rel_error.append(float(row["lc_avg_rel_error"]))

# ---------------- Runtime ----------------
plt.figure()
plt.plot(limits, mg_runtime, marker='o', label="MisraGries")
plt.plot(limits, lc_runtime, marker='o', label="Lossy Counting")
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Runtime (seconds)")
plt.title("Runtime Growth Comparison")
plt.legend()
plt.show()

# ---------------- Memory ----------------
plt.figure()
plt.plot(limits, mg_memory, marker='o', label="MisraGries")
plt.plot(limits, lc_memory, marker='o', label="Lossy Counting")
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Memory Usage (KB)")
plt.title("Memory Usage Comparison")
plt.legend()
plt.show()

# ---------------- Absolute Error ----------------
plt.figure()
plt.plot(limits, mg_abs_error, marker='o', label="MisraGries")
plt.plot(limits, lc_abs_error, marker='o', label="Lossy Counting")
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Average Absolute Error")
plt.title("Average Absolute Error (Heavy Hitters)")
plt.legend()
plt.show()

# ---------------- Relative Error ----------------
plt.figure()
plt.plot(limits, mg_rel_error, marker='o', label="MisraGries")
plt.plot(limits, lc_rel_error, marker='o', label="Lossy Counting")
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Average Relative Error (%)")
plt.title("Average Relative Error (Heavy Hitters)")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

# ---------------- READ MG + LC ----------------
limits = []

mg_runtime, mg_memory, mg_ae, mg_re = [], [], [], []
lc_runtime, lc_memory, lc_ae, lc_re = [], [], [], []

with open("metrics.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        limits.append(int(row["limit"]))

        mg_runtime.append(float(row["mg_runtime"]))
        mg_memory.append(float(row["mg_memory_kb"]))
        mg_ae.append(float(row["mg_avg_abs_error"]))
        mg_re.append(float(row["mg_avg_rel_error"]))

        lc_runtime.append(float(row["lc_runtime"]))
        lc_memory.append(float(row["lc_memory_kb"]))
        lc_ae.append(float(row["lc_avg_abs_error"]))
        lc_re.append(float(row["lc_avg_rel_error"]))

# ---------------- READ CMS ----------------
cms_runtime, cms_memory, cms_ae, cms_re = [], [], [], []

with open("cms.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        cms_runtime.append(float(row["cms_runtime"]))
        cms_memory.append(float(row["cms_memory"]))
        cms_ae.append(float(row["cms_ae"]))
        cms_re.append(float(row["cms_re"]))

# ---------------- RUNTIME ----------------
plt.figure()
plt.plot(limits, mg_runtime, marker='o', label="MisraGries")
plt.plot(limits, lc_runtime, marker='o', label="Lossy Counting")
plt.plot(limits, cms_runtime, marker='o', label="Count-Min Sketch")
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Runtime (seconds)")
plt.title("Runtime Growth Comparison")
plt.legend()
plt.show()

# ---------------- MEMORY ----------------
plt.figure()
plt.plot(limits, mg_memory, marker='o', label="MisraGries")
plt.plot(limits, lc_memory, marker='o', label="Lossy Counting")
plt.plot(limits, cms_memory, marker='o', label="Count-Min Sketch")
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Memory Usage (KB)")
plt.title("Memory Usage Comparison")
plt.legend()
plt.show()

# ---------------- ABSOLUTE ERROR ----------------
plt.figure()
plt.plot(limits, mg_ae, marker='o', label="MisraGries")
plt.plot(limits, lc_ae, marker='o', label="Lossy Counting")
plt.plot(limits, cms_ae, marker='o', label="Count-Min Sketch")
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Average Absolute Error")
plt.title("Average Absolute Error (Heavy Hitters)")
plt.legend()
plt.show()

# ---------------- RELATIVE ERROR ----------------
plt.figure()
plt.plot(limits, mg_re, marker='o', label="MisraGries")
plt.plot(limits, lc_re, marker='o', label="Lossy Counting")
plt.plot(limits, cms_re, marker='o', label="Count-Min Sketch")
plt.xlabel("Limit (Number of Queries)")
plt.ylabel("Average Relative Error (%)")
plt.title("Average Relative Error (Heavy Hitters)")
plt.legend()
plt.show()
```
